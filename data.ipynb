## Data Cleaning

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df["id"]=df["id"].astype(object)
df["host_id"]=df['host_id'].astype(object)
df.describe()
```

```python
df.info()
```

```python
df["rating"].unique()
```

```python
df["rating"] = df["rating"].astype(str).str.strip()

df["rating"] = df["rating"].replace(["No rating", "New"], np.nan)

df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

# 4. Ortalamayı al ve NaN'leri ortalama ile doldur
mean_value = df["rating"].mean()
df["rating"] = df["rating"].fillna(mean_value)


```

```python
df.info()
```

### Data Analysis

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
sns.countplot(x="room_type", data=df)
plt.title("Room type Distribution")
plt.show()
```

```python
neighbourhood_count = df["neighbourhood_group"].value_counts().reset_index()
neighbourhood_count.columns = ["neighbourhood_group", "count"]
neighbourhood_count = neighbourhood_count.sort_values("count", ascending=False)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="neighbourhood_group", data=df, order=df["neighbourhood_group"].value_counts().index)
plt.title("Number of Listings by Neighborhood Group")
plt.show()
```

```python
df["neighbourhood"].value_counts().head(10).plot(kind="bar", figsize=(10,5))
plt.xlabel("Neighbourhood")
plt.ylabel("Number of Listings")
plt.title("Top 10 Neighbourhoods with Most Airbnb Listings")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room_type", y="rating", data=df)
plt.title("Rating Distribution by Room Type")
plt.show()
```

```python
reviews_df = (
    df.groupby(["neighbourhood_group", "room_type"])["number_of_reviews"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)
```

```python
reviews_df
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="neighbourhood_group", y="number_of_reviews", data=df, estimator=sum)
plt.title("Total Number of Comments by Region")
plt.show()
```

```python
rating_df = (
    df.groupby(["neighbourhood_group", "room_type"])["rating"]
      .mean()
      .sort_values(ascending=False)
      .reset_index()
)
```

```python
rating_df
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    x="neighbourhood_group",
    y="rating",
    hue="room_type",
    data=rating_df,
    palette="Set2"
)
```
